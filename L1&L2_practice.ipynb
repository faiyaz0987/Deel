```python
from sklearn.datasets import make_moons
X_moons, y_moons = make_moons(n_samples=20000, noise=0.25, random_state=53)
X_moons.shape
y_moons.shape
```

```python
from sklearn.datasets import make_circles
X_circles,y_circles = make_circles(n_samples= 20000, noise = 0.2, random_state= 42)
X_circles
y_circles
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_moons[:,0],X_moons[:,1], c = y_moons)
plt.show()
```

```python
plt.scatter(X_circles[:,0],X_circles[:,1], c = y_circles)
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(X_moons)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_moons, y_moons, test_size=0.33, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
from tensorflow.keras.layers import Input

model = Sequential([
    Input(shape=(2,)),
    Dense(512, activation = "elu"),
    Dense(256, activation = "elu"),
    Dense(128, activation = "elu"),
    Dense(1, activation= "sigmoid")
])
```

```python
model.compile(optimizer = Adam(learning_rate = 0.001), loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
history = model.fit(X_train, y_train, epochs = 20, validation_split = 0.2)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(loss)
print(accuracy)
```

```python
plt.plot(history.history['loss'], label = 'loss')
plt.plot(history.history['val_loss'], label = 'validation loss')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.regularizers import Regularizer, l1
modell1 = Sequential()
modell1.add(Input(shape=(2,)))
modell1.add(Dense(512, activation = "elu", kernel_regularizer = l1(0.001)))
modell1.add(Dense(256, activation = "elu", kernel_regularizer = l1(0.001)))
modell1.add(Dense(128, activation = "elu", kernel_regularizer = l1(0.001)))
modell1.add(Dense(1, activation= "sigmoid"))
```

```python
learning_rate = 0.001
modell1.compile(optimizer=Adam(learning_rate=learning_rate), loss = "binary_crossentropy", metrics = ['accuracy'])
```

```python
history = modell1.fit(X_train, y_train, epochs = 100, validation_split = 0.2)
```

```python
loss, accuracy = modell1.evaluate(X_test, y_test)
print(loss)
print(accuracy)
```

```python
plt.plot(history.history['loss'], label = 'loss')
plt.plot(history.history['val_loss'], label = 'validation loss')
plt.legend()
plt.show()
```

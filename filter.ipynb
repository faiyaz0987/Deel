```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load the CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()

# Select a sample image from the training set
sample_image = train_images[0]

# Convert to a TensorFlow tensor
image_tensor = tf.convert_to_tensor(sample_image, dtype=tf.float32)

# Extract individual RGB channels using slicing
red_channel_slice = image_tensor[:, :, 0]
green_channel_slice = image_tensor[:, :, 1]
blue_channel_slice = image_tensor[:, :, 2]

# Extract individual RGB channels using tf.split
red_channel_split, green_channel_split, blue_channel_split = tf.split(image_tensor, num_or_size_splits=3, axis=2)

print("Shape of original image tensor:", image_tensor.shape)
print("Shape of Red channel (slicing):", red_channel_slice.shape)
print("Shape of Green channel (slicing):", green_channel_slice.shape)
print("Shape of Blue channel (slicing):", blue_channel_slice.shape)
print("Shape of Red channel (tf.split):", red_channel_split.shape)

# You can visualize the original image and the channels
plt.figure(figsize=(10, 4))

plt.subplot(1, 4, 1)
plt.imshow(sample_image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(red_channel_slice, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(green_channel_slice, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(blue_channel_slice, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

plt.tight_layout()
plt.show()
```
